```python
from scipy.io import arff
import pysubgroup as ps
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.DataFrame (arff.loadarff("/Users/lemmerich/Documents/data/data/credit-g.arff") [0])
```

```python
target = ps.NumericTarget ('age')
searchSpace = ps.createSelectors(data, ignore=['age'], nbins=10)

task = ps.SubgroupDiscoveryTask (data, target, searchSpace, 
                                 resultSetSize=10, depth=3, 
                                 qf=ps.StandardQF_numeric(1, False))
```

```python
%time result = ps.DFS_numeric().execute(task)
for (q, sg) in result:
    print (str(q) + ":\t" + str(sg.subgroupDescription))
```

```python
ps.as_df (data, result, ps.all_statistics_numeric)
```

```python
task = ps.SubgroupDiscoveryTask (data, target, searchSpace, 
                                 resultSetSize=5, depth=2, 
                                 qf=ps.GAStandardQF_numeric(1, False))


%time result = ps.SimpleDFS().execute(task)
ps.as_df (data, result, ps.all_statistics_numeric)
```

```python
sg = result[0][1]
fig = ps.plot_distribution_numeric(sg, data, bins = np.linspace(18, 90, 20))
```
